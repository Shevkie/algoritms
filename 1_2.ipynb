```python
# импорт библиотек
import random
import math
```

```python
# функция
def f(x):
    f_x = (math.sin(x))/x
    return f_x
```

```python
# метод монте карло: 
def IntMonte_Karlo(a, b, N , function):
    # a - нижняя граница
    # b - верхняя граница
    # N - кол-во случайных точек
    # h - шаг
    h = (b - a) / N
    i = 1
    IntSumma = 0
    while (i <= N):
        IntSumma += function(random.uniform(a,b)) #накапливает сумму значений функции в выбранных случайных точках
        i += 1
    IntResult = h * IntSumma #произведение шага h на сумму значений функции
    print("Рузультат:", IntResult)
    print("Погрешность:",   abs(IntResult - 1.5773067) / 1.5773067 * 100)
    return 
```

```python
IntMonte_Karlo(0.001,11,1000000, f)
```
